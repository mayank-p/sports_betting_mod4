```python
def get_season_stats(tid):
    team_season = leaguegamefinder.LeagueGameFinder(team_id_nullable=tid).get_data_frames()[0]
    team_season = team_season[team_season.GAME_DATE >= '2019-10-22']
    return team_season
    
```
