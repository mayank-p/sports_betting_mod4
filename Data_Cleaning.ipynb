### Create Cleaning Function Following Same Steps

```python
def clean_data(csv_file):
    df = pd.read_csv(csv_file)
    
    #set teams
    df['Team1'] = df['Match\xa0Up'].str.split().apply(lambda x: x[0])
    df['Team2'] = df['Match\xa0Up'].str.split().apply(lambda x: x[2])
    
    #delete unused columns
    df.drop(['Unnamed: 0','Season','Match\xa0Up','Game\xa0Date','W/L','MIN'], axis = 1,inplace = True)
    
    #set targets
    df['win_target'] = df['+/-'] > 0
    df['3_win_target'] = df['+/-'] > 3
    df['5_win_target'] = df['+/-'] > 5
    df['7_win_target'] = df['+/-'] > 7
    df['10_win_target'] = df['+/-'] > 10
    
    #drop last unused column
    df.drop(['+/-'],axis = 1,inplace=True)
    
    return df
```

```python
clean_df = clean_data('nba_games.csv')
```
